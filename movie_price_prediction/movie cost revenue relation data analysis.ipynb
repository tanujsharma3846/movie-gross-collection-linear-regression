```python
import pandas as pd
from pandas import DataFrame
import  matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("cost_revenue_clean.csv")
data
```

```python
data.describe()
```

```python
3.29e7
```

```python
x=DataFrame(data,columns=['production_budget_usd'])
y=DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title("film cost vs gloval revenue")
plt.xlabel("production budget $")
plt.ylabel("world wide gross $")
plt.ylim(0, 3000000000)# 0 t0 3 billion
plt.xlim(0, 450000000)# 0 to 450 million
```

```python
regression=LinearRegression()
regression.fit(x,y)
```

slope coefficient

```python
regression.coef_
```

```python
#intercept
regression.intercept_
```

```python
# plotting linear regression line
```

```python

fig, ax = plt.subplots(figsize=(10, 8))

ax.scatter(x, y, s=40, alpha=0.3)


ax.plot(x.values, regression.predict(x.values), color="red", linewidth=4)

plt.title("Film Cost vs Global Revenue")

plt.xlabel("Production Budget $")
plt.ylabel("Worldwide Gross $")

plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)

plt.show()
```
